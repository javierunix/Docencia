- we have to applay the following formula: $c(g/L) = \frac{mass\ of\ solute\ (g)}{volume\ of\ solution\ (L)}$

```python
mass = 8
volume = 400/1000

def c_g_L(mass, volume):
    return mass/volume

c_g_L(mass, volume)
```

2. Calculate the concentration of a solution prepared by dissolving 25 g of sugar in 375 g of water

- the formula to apply is $\%\ mass = \frac{mass\ of\ solute}{mass\ of\ solute\ + mass\ of\ solvent} \times 100$

```python
mass_solute = 25
mass_solvent = 375

def c_mass(mass_solute, mass_solvent):
  return mass_solute / (mass_solute + mass_solvent) * 100

c_mass(mass_solute, mass_solvent)
```

3. Calculate the % volume of nitrogen of 200 L of a solution of air if it contains 156 L of nitrogen

- the formula to apply is $\%\ volume = \frac{volume\ of\ solute}{volume\ of\ solution} \times 100$

```python
volume_solute = 156
volume_solution = 200

def c_vol(volume_solute, volume_solution):
  return volume_solute / volume_solution * 100

c_vol(volume_solute, volume_solution)
```
