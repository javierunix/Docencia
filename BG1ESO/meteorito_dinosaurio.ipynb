# Energía Cinética

La energía que tiene un cuerpo debido a su movimiento se denomina energía cinética, y tiene la siguiente fórmula $Ec = \frac{m\ \times V^{2}}{2}$

- m es la masa del cuerpo (en este caso el meteorito)
- v es la velocidad del cuerpo

Se supone que la velocidad del meteorito era de unos 20 km/h. Dado que debemos utilizar en nuestros cálculos las unidades del sistema internacional, debemos expresarla en m/s.

**Por lo tanto, podemos concluir que v = 20000 m/s**.

Queda por averiguar la masa del meteorito.

## La masa del meteorito

Para estimar la masa del meteorito hay que hacer primero una estimación de cuál podía ser su tamaño. Se piensa que su tamaño podía ser de unos 11 km en su eje mayor.

Haciendo un estimación conservadora, vamos a suponer que sus ejes menores eran de 4 km, cada uno.

En ese caso el volumen sería de $11\ \times\ 4\ \times\ 4\ = 176\ km^3 $

Otro dato necesario es el de la densidad. La mayoría de los meteoritos tienen densidades que se situan entre los 3 y los 4 $g/cm^3$ (nosotros vamos a considerar el término medio de este rango).

Como hicimos, antes todas las magnitudes hay que expresarlas en unidades del SI.

- volumen= $176 \times 10^9\ m^3$
- densidad = $3500\ kg/m^3$

## Estimación de la energía cinética del impacto

Una vez que hemos estimado todas las magnitudes necesarias en las unidades del SI podemos hacer los cálculos.


```python
# estimación de la masa del meteorito
v = 20000 # velocidad del meteorito
vol = 176000000000 # volumen del meteorito
d = 3500 # densidad del meteorito
mass = vol * d # cálculo de la masa
ec = 1/2 * mass * (v**2) # cálculo de la energía cinética
ec
```
